# creating popularity based recommendation system 
idea is to recommend top 100 highest average ratings book

```python
books_ratings = rating.merge(books, on='ISBN')
```

```python
books_ratings.columns
```

```python
rating_df = books_ratings.groupby('Book-Title').count()['Book-Rating'].reset_index()
rating_df = rating_df.rename(columns={'Book-Rating': 'total_ratings'})
```

```python
avg_rating_df = books_ratings.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df = avg_rating_df.rename(columns={'Book-Rating': 'avg_ratings'})
```

```python
popularity_df = rating_df.merge(avg_rating_df, on='Book-Title')
popularity_df
```

```python
popularity_df = popularity_df[popularity_df['total_ratings'] >= 250].sort_values('avg_ratings', ascending=False).head(25)
```

```python
popularity_df
```

```python
df = popularity_df.merge(books, on ='Book-Title')
df = df.drop_duplicates('Book-Title')
```

```python
df.columns
```

```python
df = df[['Book-Title', 'Book-Author', 'Year-Of-Publication', 'Image-URL-M', 'total_ratings', 'avg_ratings']]
df
```

# Creating Collaborative Recommender System

```python
x= books_ratings.groupby('User-ID').count()['Book-Rating']>200
rated_users =  x[x].index
```

```python
filtered_ratings= books_ratings[books_ratings['User-ID'].isin(rated_users)]
filtered_ratings
```

```python
y=filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=50
y[y]
```

```python
famous_books = y[y].index
famous_books
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings.duplicated().sum()
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0, inplace = True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity_score = cosine_similarity(pt)
```

```python
#fetch index from the book name

def recommend(book_name):
    index = np.where(pt.index == book_name)[0][0]
    similar_books = sorted(list(enumerate(similarity_score[index])), key=lambda x:x[1], reverse=True)[1:20]
    data =[]
    
    for i in similar_books:
        item =[]
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
        
    return data
#     return suggestions
```

```python
recommend('The Notebook')
```
